```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

MACHO

```python
eficiencia_macho= pd.read_csv("eficiency_MACHO.txt", decimal='.', sep=';')
eficiencia_macho = eficiencia_macho.values

plt.figure(figsize=(8,6))
ax=plt.axes()
plt.semilogx(eficiencia_macho[:,3],eficiencia_macho[:,4],color='black',linestyle='-',marker='o',lw=3,markersize=9)
plt.ylabel('Eficiencia MACHO',fontsize=30)
plt.xlabel(r'$\hat{t}$ [d√≠as]',fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(0,0.5)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=9)
ax.tick_params(which='minor', length=5, color='black')
ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=9)
ax.tick_params(which='minor', length=5, color='black')



```

```python
eficiencia_hsc= pd.read_csv("eficiency_HSC.txt", decimal='.', sep=';')
eficiencia_hsc = eficiencia_hsc.values


plt.figure(figsize=(8,6))
ax=plt.axes()
plt.ylim(0,0.7)
plt.ylabel('Eficiencia HSC',fontsize=30)
plt.xlabel(r'$\hat{t}$ [horas]',fontsize=30)
plt.axvline(x=1/60*3,ymin=0,ymax=1,color='orange')
plt.axvline(x=4,ymin=0,ymax=1,color='orange')
plt.axhline(y=0.55,xmin=0,xmax=1,color='red')
plt.axvspan(1/60*3, 4, alpha=0.3, color='y')
plt.semilogx(eficiencia_hsc[:,0],eficiencia_hsc[:,1],color='black',linestyle='-',marker='o',lw=3,markersize=9)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=9)
ax.tick_params(which='minor', length=5, color='black')

print(1/60*3)
print(eficiencia_hsc[:,0][0],eficiencia_hsc[:,0][-1])
```
